```python
df = pd.read_csv("/content/drive/MyDrive/resource/price_prediction2.csv")
df
```

```python
print(df.columns)
```

```python
%matplotlib inline
plt.ylabel("price")
plt.xlabel("area")
plt.scatter(df.area , df.price , color='red', marker='+')
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
d= pd.read_csv("/content/drive/MyDrive/resource/area.csv")
d
```

```python
print(df.columns)
print(d.columns)
```

```python
p= reg.predict(d)
```

```python
d['prices']=p
```
